**TIPS DATASET**


```python
import seaborn as sns
import matplotlib.pyplot as plt
tips=sns.load_dataset("tips")
sns.set(style="whitegrid")
sns.scatterplot(data=tips,x="total_bill",y="tip",hue="time",style="time",markers=["o",'D'])
plt.title("Tips Dataset",fontsize=15)
plt.xlabel("Total bill($)")
plt.ylabel("Tip ($)")
plt.legend(title="Time of the Day",loc="upper right")
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
iris=sns.load_dataset("iris")
sns.set(style="whitegrid")
sns.scatterplot(data=iris,x="sepal_length",y="sepal_width",hue="species",style="species",markers=["o","D","*"])
plt.title("IRIS DATASET",fontweight="bold")
plt.xlabel("Sepal Length",fontweight="bold")
plt.ylabel("Sepal Width",fontweight="bold")
plt.legend(title="Species",loc="upper right")
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
titanic=sns.load_dataset("titanic")
sns.set(style="whitegrid")
sns.scatterplot(data=titanic,x="fare",y="age",hue="class",style="class",markers=["o","D","."])
plt.title("TITANIC DATASET",fontweight="bold")
plt.xlabel("Fare",fontweight="bold")
plt.ylabel("Age",fontweight="bold")
plt.legend(title="CLASS",loc="upper right")
plt.show()
```

```python
#HEATMAP
import seaborn as sns
flights=sns.load_dataset("flights")
plt.figure(figsize=(12,6))
sns.heatmap(flights.pivot_table(index="month", columns="year", values="passengers"),cmap="YlGnBu",annot=True,fmt='.0f')
plt.title("Flights data set -Heatmap")
plt.show()
```

```python
#LINE PLOT 
import seaborn as sns
import matplotlib.pyplot as plt
flights=sns.load_dataset("flights")
plt.figure(figsize=(12,6))
sns.lineplot(data=flights,x="year",y="passengers",hue="month",marker="o",linestyle="solid")
plt.legend(title="Month")
plt.xlabel("Year")
plt.ylabel("Passengers")
plt.grid(False)
plt.show()
```

```python
#BARPLOT
plt.figure(figsize=(12,6))
sns.barplot(data=flights,x="month",y="passengers",ci=None)
plt.title('Flights Dataset - Average Passengers Per Month')
plt.xlabel('Month')
plt.ylabel('Average Passengers')
plt.show()
```

**Boxplot**

sns.boxplot is a Seaborn function used to create box plots. Box plots are a type of categorical plot that display the distribution of a numerical variable across different categories. They are useful for visualizing the spread, central tendency, and potential outliers in the data.

```python
tips=sns.load_dataset("tips")
sns.boxplot(x="day",y="total_bill",data=tips)
```

`Difference between bar plot and box plot sns.barplot:`
Purpose: sns.barplot is used to display the central tendency (commonly the mean) of a numerical variable for different categories. It shows the average value and confidence interval (by default, with a bootstrapped confidence interval) for each category.

Visualization: The main plot consists of vertical bars, where the height of each bar represents the mean value of the numerical variable for a specific category.

Use Case: It is suitable for comparing the average values across categories and identifying differences in central tendency.

sns.boxplot:

Purpose: sns.boxplot is used to display the distribution, spread, and presence of outliers of a numerical variable for different categories. It shows the quartiles, median, and potential outliers for each category.

Visualization: The main plot consists of boxes (interquartile range, IQR) with lines (whiskers) extending to show the range of the data. Outliers are typically shown as individual points.

Use Case: It is suitable for comparing the spread and distribution of the numerical variable across categories, identifying potential outliers, and assessing the overall variability.

**Violin Plots**

A violin plot is a type of data visualization that combines aspects of a box plot and a kernel density plot. It is particularly useful for visualizing the distribution of a numerical variable or the comparison of multiple numerical variables across different categories.

```python
sns.violinplot(x="day",y="total_bill",data=tips)
```

**LM Plot**

The lmplot function in Seaborn is used to create scatter plots with overlaid linear regression models. It is particularly useful for visualizing the relationship between two numerical variables and fitting a linear regression model to the data. The term "lmplot" stands for "linear model plot."

```python
sns.lmplot(data=tips,x="total_bill",y="tip",hue="day")
```
