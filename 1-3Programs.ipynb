4 a) Write a Python program to Demonstrate how to Draw a Bar Plot using Matplotlib.

```python
'''BAR PLOT'''
import matplotlib.pyplot as plt
x=["R", "c" ,"c++", "java", "python"]
y=[50 60 75 80 95]
plt.title("Programming languages Popularity" ,fontsize=10)
plt.xlabel("Programming languages" ,fontsize=10)
plt.ylabel("Popularity")
c=["r", "b", "g", "pink", "purple"]
plt.bar(x, y ,color=c ,width=0.5, edgecolor="r", linewidth=3, alpha=0.2, label="languages popularity")
plt.legend(loc="upper left")
plt.xticks(x ,rotation=90)
plt.grid()
plt.show()
```
