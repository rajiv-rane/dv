### In a histogram  the data is divided into intervals  or "bins " and the number of data points falling into each bin is represented by the height of a bar.

```python
import matplotlib.pyplot as plt
import numpy as np
import random
x=np.random.randint(10,60,20)
print(x)
```

```python
x=[12 ,13 ,19, 13, 53, 51, 49, 35, 34 ,52, 23, 25, 33, 14, 47, 57, 43 ,43 ,45,17]
plt.hist(x,edgecolor="green",label="marks",color="magenta")
plt.legend()
plt.title("Histogram Chart")
plt.xlabel("Python")
plt.ylabel("Number")
plt.show()
```

```python
#using bins
import matplotlib.pyplot as plt
x=[47,13,52,16,15,58,57,32,35,13,56,46,42,47,37,23,33,50,32,14,23,35,52,16,51,
    47,22,22,16,24,54,13,16,45,39,58,38,19,58,25,25,25,40,44,10,42,50,23,35,27]
plt.hist(x,color="green",bins=10,label="marks",edgecolor="red") #bins divides the data into given number of intervals/graphs
plt.title("Histogram Chart")
plt.xlabel("Python")
plt.ylabel("No.")
plt.show()
```

```python
#reverse order/decreasing or increasing histogram
'''for decreasing=> cumulative=-1
for increasing=> cumulative=1'''
import matplotlib.pyplot as plt
x=[47,13,52,16,15,58,57,32,35,13,56,46,42,47,37,23,33,50,32,14,23,35,52,16,51,
    47,22,22,16,24,54,13,16,45,39,58,38,19,58,25,25,25,40,44,10,42,50,23,35,27]
plt.hist(x,color="pink",alpha=0.5,edgecolor="r",label="No",cumulative=1)
plt.title("Histogram")
plt.xlabel("Python")
plt.ylabel("No")
plt.legend()
plt.show()
```

```python
'''ALIGN the values in mid or edge for histogram'''
#bottom=10 means the y axis values begin from 10
import matplotlib.pyplot as plt
x=[47,13,52,16,15,58,57,32,35,13,56,46,42,47,37,23,33,50,32,14,23,35,52,16,51,
    47,22,22,16,24,54,13,16,45,39,58,38,19,58,25,25,25,40,44,10,42,50,23,35,27]
plt.hist(x,color="yellow",alpha=0.8,bottom=10,edgecolor="green",align="mid")
plt.show()
```

```python
#Differentiate b/w hist and bar plot 
#stepfilled => No edges visible

import matplotlib.pyplot as plt
x=[47,13,52,16,15,58,57,32,35,13,56,46,42,47,37,23,33,50,32,14,23,35,52,16,51,
    47,22,22,16,24,54,13,16,45,39,58,38,19,58,25,25,25,40,44,10,42,50,23,35,27]
plt.hist(x,color="yellow",alpha=0.8,bottom=10,edgecolor="green",align="mid",histtype="stepfilled")
plt.show()
```

```python
'''orientation= "vertical" or "horizontal"'''
import matplotlib.pyplot as plt
x=[47,13,52,16,15,58,57,32,35,13,56,46,42,47,37,23,33,50,32,14,23,35,52,16,51,
    47,22,22,16,24,54,13,16,45,39,58,38,19,58,25,25,25,40,44,10,42,50,23,35,27]
plt.hist(x,color="yellow",alpha=0.8,bottom=10,edgecolor="green",align="mid",orientation="horizontal")
plt.show()
```

```python
#rwidth =>optional parameter and it is a relative width of the bars as a fraction of the bin width
import matplotlib.pyplot as plt
x=[47,13,52,16,15,58,57,32,35,13,56,46,42,47,37,23,33,50,32,14,23,35,52,16,51,
    47,22,22,16,24,54,13,16,45,39,58,38,19,58,25,25,25,40,44,10,42,50,23,35,27]
plt.hist(x,color="pink",alpha=0.5,edgecolor="r",label="No",rwidth=0.8)
plt.title("Histogram")
plt.xlabel("Python")
plt.ylabel("No")
plt.legend()
plt.show()
```

```python
# histogram chart for logarithmic table 
import matplotlib.pyplot as plt
x=[47,13,52,16,15,58,57,32,35,13,56,46,42,47,37,23,33,50,32,14,23,35,52,16,51,
    47,22,22,16,24,54,13,16,45,39,58,38,19,58,25,25,25,40,44,10,42,50,23,35,27]
plt.hist(x,color="pink",alpha=0.5,edgecolor="r",label="No",log=True)
plt.title("Histogram")
plt.xlabel("Python")
plt.ylabel("No")
plt.legend()
plt.show()
```

```python
#vertical line to divide histogram chart & grid
import matplotlib.pyplot as plt
x=[47,13,52,16,15,58,57,32,35,13,56,46,42,47,37,23,33,50,32,14,23,35,52,16,51,
    47,22,22,16,24,54,13,16,45,39,58,38,19,58,25,25,25,40,44,10,42,50,23,35,27]
plt.hist(x,color="green",alpha=0.5,edgecolor="r",label="No",rwidth=0.8)
plt.axvline(30,color="yellow",label="partition")
plt.title("Histogram")
plt.xlabel("Python")
plt.ylabel("No")
plt.legend()
plt.grid()
plt.show()
```

**5(A) PIECHART**



```python
import matplotlib.pyplot as plt 
x=[10, 20,30,40]
y=["C++","Java","Python","Js"]
plt.pie(x,labels=y)
plt.show()
```

```python
#explode => separated from the pie chart
import matplotlib.pyplot as plt
x=[10,20,30,40]
y=["C","C#","JS","SQL"]
ex=[0.5,0,0,0]
c=["r","b","g","y"] #changing default colors
plt.pie(x,labels=y,explode=ex,colors=c)
plt.show()
```

```python
#adding % symbol inside the piechart , SHADOW
# autopct=%0.1f%%
import matplotlib.pyplot as plt
x=[10,20,30,40]
y=["C","C#","JS","SQL"]
ex=[0.5,0,0,0]
c=["r","b","g","y"] #changing default colors
plt.pie(x,labels=y,explode=ex,colors=c,autopct="%0.1f%%",shadow=True,radius=0.5,labeldistance=1.2)
plt.show()
```

```python
#ROTATE 90 degree COUNTER CLOCKWISE => startangle=90
import matplotlib.pyplot as plt
import matplotlib.pyplot as plt
x=[10,20,30,40]
y=["C","C#","JS","SQL"]
ex=[0.5,0,0,0]
c=['r','b','g','y']
plt.pie(x,labels=y,colors=c,explode=ex,labeldistance=1.5,startangle=90,radius=0.9,autopct="%0.1f%%",shadow=True)

plt.show()
```

```python
# Rotate clock wise by counterclock=False
import matplotlib.pyplot as plt 
x=[10,20,30,40]
y=["C","C#","JS","SQL"]
ex=[0.5,0,0,0]
c=['r','g','b','y']
plt.pie(x,labels=y,colors=c,explode=ex,startangle=90,textprops={"fontsize":15},counterclock=False,shadow=True)
plt.show()
```

```python
#adding linewidth,edgecolor in dict using wedgeprops={}
#rotatelabels=angle or True
x=[10,20,30,40]
y=["C","C#","JS","SQL"]
ex=[0.5,0,0,0]
c=['r','g','b','y']
plt.pie(x,labels=y,radius=0.9,shadow=True,textprops={"fontsize":12},wedgeprops={"linewidth":3,"edgecolor":"m"},rotatelabels=True)
plt.title("PIE chart")
plt.legend(loc=2)
plt.show()
```

```python
#DOUBLE PIECHART
import matplotlib.pyplot as plt
x=[10,20,30,40]
x1=[40,30,20,10]
y=["c","c++","Java","Python"]
c=["r","g","b","y"]
plt.pie(x,labels=y,colors=c,radius=1.5)
plt.pie(x1,labels=y,radius=1)
plt.show()
```

```python
#Donut piechart
import matplotlib.pyplot as plt
x=[10,20,30,40]
y=["c","c++","Java","Python"]
plt.pie(x,labels=y,radius=1.5)
plt.pie([1],colors="w",radius=1)
plt.show()
```
